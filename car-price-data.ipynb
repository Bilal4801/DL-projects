```python
data1 = car_pr[['horsepower', 'drivewheel', 'wheelbase', 'boreratio', 'enginesize', 'curbweight',
               'carwidth', 'carlength', 'price', 'enginelocation']]
```

# Split data

```python
x = (data1.loc[:, data1.columns != 'price'])
y = (data1.loc[:, data1.columns == 'price'])

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30, random_state = 42)
```

```python
x_train2, x_val, y_train2, y_val = train_test_split(x_train, y_train, test_size=0.1, random_state=42)
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
def built_model():
    model1 = models.Sequential()
    model1.add(layers.Dense(30, activation='relu', input_shape=(x_train2.shape[1],)))
    model1.add(layers.Dropout(0.5))
    model1.add(layers.Dense(1))
    
    model1.compile(optimizer='rmsprop', loss='mse', metrics='mae')
              
    return model1
```

```python
model3 = built_model()
```

```python
history1 = model3.fit(x_train2, y_train2, epochs=50,
                   validation_data=(x_val, y_val))
history1
```

```python
test_loss, test_accuracy = model3.evaluate(x_test, y_test)
```
