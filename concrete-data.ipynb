# split data

```python
from sklearn.model_selection import train_test_split
```

```python
X = conc[['Cement (component 1)(kg in a m^3 mixture)',
          'Blast Furnace Slag (component 2)(kg in a m^3 mixture)',
          'Fly Ash (component 3)(kg in a m^3 mixture)',
          'Water  (component 4)(kg in a m^3 mixture)',
          'Superplasticizer (component 5)(kg in a m^3 mixture)',
          'Coarse Aggregate  (component 6)(kg in a m^3 mixture)',
          'Fine Aggregate (component 7)(kg in a m^3 mixture)', 'Age (day)']]
```

```python
y = conc[['Concrete compressive strength(MPa, megapascals) ']]
```

```python
print("shape of X is =", X.shape)
print('shape of y is =',y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=51)
```

```python
x_train1, x_val, y_train1, y_val = train_test_split(x_train, y_train, test_size=0.1, random_state=51)
```

```python
print("shape of X is =", x_train.shape)
print('shape of y is =', y_train.shape)
print("shape of X is =", x_test.shape)
print('shape of y is =', y_test.shape)
```

# Strandradize the input variables

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train1)
x_test_sc = sc.fit_transform(x_test)
x_val_sc = sc.fit_transform(x_val)
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
def built_model2():
    model = models.Sequential()
    model.add(layers.Dense(10, activation='tanh', input_shape=(8,)))
    model.add(layers.Dense(8, activation='relu'))
    model.add(layers.Dense(1))
    
    model.compile(optimizer='rmsprop', loss='mse', metrics='mae')
    
    return model
```

```python
model = built_model2()
data = model.fit(x_train_sc, y_train1, epochs=50, validation_data = (x_val_sc, y_val))
data
```

```python
test_loss, test_accu = model.evaluate(x_test_sc, y_test)
```
