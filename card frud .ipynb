```python
train_data = train.drop(columns='Class')
```

```python
train_data.head()
```

```python
test_data = test.drop(columns='Class')
```

```python
test_data.head()
```

```python
test_data.dtypes
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
def built_model5():
    model2 = models.Sequential()
    model2.add(layers.Dense(100, activation='relu', input_shape=(train_data.shape[1],)))
    model2.add(layers.Dropout(0.5))
    model2.add(layers.Dense(1, activation='sigmoid'))
    
    model2.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
    
    return model2
```

```python
model4 = built_model5()
```

```python
history = model4.fit(train_data, train_label, epochs=10, validation_split=0.3,
                    validation_batch_size = 50, validation_steps = 10)
```

```python
# len(df) * 2/len(train_data)
```

```python
test_loss, test_accuracy = model4.evaluate(test_data, test_label)
```
